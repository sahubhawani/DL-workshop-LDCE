```python
#Detecting number plates

from PIL import Image
import pytesseract

count = 0
number = []
num_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_russian_plate_number.xml")

cap = cv2.VideoCapture("/home/sagar/plate_test_1.mp4")

while True:
    success , img = cap.read()
    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    num_plates = num_cascade.detectMultiScale(gray_img , 1.1,4)

    for (x , y, w , h) in num_plates:
        area = w * h
        if area > 250:
            cv2.rectangle(img, (x,y),(x+w,y+h),(255,0,0),2)
            cv2.putText(img, "Number Plate",(x,y-5),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(255,0,0),2)
            imgROI = img[y:y+h,x:x+w]
            cv2.imshow("ROI",imgROI)
            

    cv2.imshow("Video",img)
    
    if cv2.waitKey(1) & 0xff ==ord("s"):
        cv2.imwrite("/home/sagar/Num_plate_detected"+str(count)+".jpg",imgROI)
        print("Scanned image")
        text = pytesseract.image_to_string("/home/sagar/Num_plate_detected"+str(count)+".jpg")
        print(text)
        number.append(text)
        count += 1
    
    if cv2.waitKey(1) & 0xff ==ord("q"):
        cv2.destroyAllWindows()
        break
```
